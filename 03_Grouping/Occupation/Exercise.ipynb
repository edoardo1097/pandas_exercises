### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
df_users= pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
```

```python
df_users
```

### Step 4. Discover what is the mean age per occupation

```python
df_occupation = df_users.groupby("occupation")
```

```python
df_occupation.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df_users.head(2)
```

```python
def male(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
```

```python
df_users['Male'] = df_users['gender'].apply(male)
```

```python
df_users.head(3)
```

```python
m = df_users.groupby('occupation').Male.sum() / df_users.occupation.value_counts() * 100 
m.head(2)
```

```python
m.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df_occupation['age'].agg([max, min])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df_og = df_users.groupby(['occupation', 'gender'])
df_og.age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
df_og2 = df_og.agg({'gender': 'count'})
```

```python
df_og2
```

```python
df_occ = df_users.groupby(['occupation']).agg('count')
df_occ.head(2)
```

```python
gender_occ = df_og2.div(df_occ, level = "occupation") * 100
gender_occ.loc[: , 'gender']
```
